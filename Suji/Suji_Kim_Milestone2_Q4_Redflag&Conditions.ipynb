### Data Preprocessing

```python
import re

def clean_case_name(text):
    if pd.isna(text): return ""
    text = text.lower()
    # Explicitly handle 'vs' to 'v' for legal consistency
    text = text.replace(" vs ", " v ")
    # FIX: changed 0-0 to 0-9
    text = re.sub(r'[^a-z0-9\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

# Apply cleaning
fair_use_cases['clean_key'] = fair_use_cases['case'].apply(clean_case_name)

fair_use_findings['combined_name'] = fair_use_findings['title'].astype(str) + " " + fair_use_findings['case_number'].astype(str)
fair_use_findings['clean_key'] = fair_use_findings['combined_name'].apply(clean_case_name)

fair_use_merged_v2 = fair_use_cases.merge(
    fair_use_findings.drop(columns=["title", "case_number", "year", "court", "outcome"]),
    on="clean_key",
    how="left"
)

joined_v2 = fair_use_merged_v2["key_facts"].notna().sum()
total_v2 = len(fair_use_merged_v2)
print(f"Improved Joined Rows: {joined_v2} / {total_v2} ({joined_v2/total_v2:.1%})")
```

```python
fair_use_merged_v2
```

```python
# Outlier Indicator
df_final = fair_use_merged_v2[fair_use_merged_v2['key_facts'].notna()].copy()
df_final['holding_word_count'] = df_final['holding'].fillna('').apply(lambda x: len(x.split()))
# Label Complexity
df_final['category_count'] = df_final['categories'].fillna('').apply(lambda x: len(x.split(';')) if x else 0)

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.boxplot(y=df_final['holding_word_count'], color='salmon')
plt.title('Identifying Length Outliers in Holdings')

plt.subplot(1, 2, 2)
sns.histplot(df_final['category_count'], bins=10, kde=False, color='skyblue')
plt.title('Distribution of Category Complexity')
plt.show()
```

The graph revealed a **"long-tail distribution,"** showing that most events are straightforward (concise), but a very small number of events are extremely long and complex.

This distribution provides visual evidence for "why we should focus on outliers (red flags) rather than the average."

## 3. Analysis & Experiments

### 3.1 Statistical Outlier Detection (Isolation Forest)

```python
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler

# Selecting features to use for analysis (length, number of categories, year)
features = df_final[['holding_word_count', 'category_count', 'year']].fillna(0)
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Applying the Isolation Forest model (setting the top 5% as outliers)
model = IsolationForest(contamination=0.05, random_state=42)
df_final['is_outlier'] = model.fit_predict(features_scaled)

# Output the list of 'red flags' found by the model
red_flag_list = df_final[df_final['is_outlier'] == -1]
print(f"Found {len(red_flag_list)} Red-flag cases.")
red_flag_list[['case', 'holding_word_count', 'category_count']]
```

Finding 9 out of 177 (approximately 5%) is considered a very reasonable proportion. These are defined as **'risk groups that need careful examination'**.
These are not simply cases with long texts, but rather cases that are unusual due to a combination of length, complexity, and year.

### 3.2 Semantic Outlier Detection (PCA on Text)

Initially, we used K-Means, but the similarity of legal terminology caused outliers to be overlooked. Therefore, I switched to Isolation Forest.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import PCA

# TF-IDF - Implementing the 'Unsupervised NLP' mentioned in section 1.3 in practice
tfidf = TfidfVectorizer(stop_words='english', max_features=1000)
tfidf_matrix = tfidf.fit_transform(df_final['holding'].fillna(''))

# Two-dimensional dimensionality reduction using PCA
pca = PCA(n_components=2, random_state=42)
pca_results = pca.fit_transform(tfidf_matrix.toarray())
df_final['pca_1'] = pca_results[:, 0]
df_final['pca_2'] = pca_results[:, 1]

# Normal cases vs. Red-flag cases (found by Isolation Forest)
plt.figure(figsize=(10, 7))
sns.scatterplot(
    x='pca_1', y='pca_2',
    hue='is_outlier',
    style='is_outlier',
    palette={1: 'grey', -1: 'red'},
    alpha=0.6,
    data=df_final
)
plt.title('Semantic Landscape: Normal vs. Red-flag Cases (PCA)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Outlier Status', labels=['Normal (1)', 'Red-flag (-1)'])
plt.show()
```

If the red dots, identified as statistical outliers by the Isolation Forest algorithm, are also scattered on the periphery in the semantic (PCA) space, it proves that "these events are not only lengthy but also use unusual vocabulary (themes)."

However, if the red dots are clustered in the center, it provides a counterintuitive insight: "These are events with predictable content but excessively long descriptions."

```python
import numpy as np

explained_variance = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance)

print(f"PC1 explains: {explained_variance[0]:.2%} of the variance")
print(f"PC2 explains: {explained_variance[1]:.2%} of the variance")
print(f"Total variance explained: {cumulative_variance[-1]:.2%}")

plt.figure(figsize=(6, 4))
plt.bar(['PC1', 'PC2'], explained_variance, color='black', alpha=0.7)
plt.title('Variance Explained by Components')
plt.ylabel('Variance Ratio')
plt.show()
```

**Technical Validation of PCA:** The PCA results show that the first two principal components explain 4.88% of the total variance. While this number may appear low in traditional tabular data, it is a standard observation in high-dimensional NLP tasks where semantic meaning is highly distributed across a sparse vocabulary of 1,000 features. This low concentration of variance further justifies the use of Isolation Forest, as it confirms that the 'Red-flags' are not defined by a few simple keywords, but by complex, multi-dimensional patterns that cannot be fully captured in a 2D plane.

### 3.3 3D Semantic Mapping & Feature Loadings

While the 2D PCA provided a baseline, the low explained variance (4.88%) suggested that the semantic nuances are too complex for a single plane. I initiated a 3D Semantic Mapping with Loading Vectors to achieve two goals:

* To provide a **higher-resolution view of the data by adding a third principal component**.

* To transition from mere **observation** to **interpretation** by projecting specific keywords (loadings) into the space. This allows me to see which legal themes (e.g., parody vs. photography) are 'pulling' certain Red-flag cases away from the normal cluster."

```python
import plotly.express as px
import plotly.graph_objects as go
import numpy as np

# PCA with 3 components
pca_3d = PCA(n_components=3, random_state=42)
pca_results_3d = pca_3d.fit_transform(tfidf_matrix.toarray())

df_final['pc1'] = pca_results_3d[:, 0]
df_final['pc2'] = pca_results_3d[:, 1]
df_final['pc3'] = pca_results_3d[:, 2]

# Identify the top words
feature_names = tfidf.get_feature_names_out()
loadings = pca_3d.components_.T * np.sqrt(pca_3d.explained_variance_)

# Indices of the top 10 words that contribute most to the variance
top_features_idx = np.argsort(np.abs(loadings).sum(axis=1))[-10:]


fig = px.scatter_3d(
    df_final, x='pc1', y='pc2', z='pc3',
    color='is_outlier',
    hover_name='case',
    hover_data=['year', 'holding_word_count'],
    title='3D Semantic Landscape of Fair Use Cases',
    # Red(-1) / Grey(1))
    color_continuous_scale=['red', 'grey'],
    opacity=0.7
)

# Add top 10 word arrows
scale_factor = pca_results_3d.max() * 0.8

for i in top_features_idx:
    word = feature_names[i]
    # Line
    fig.add_trace(go.Scatter3d(
        x=[0, loadings[i, 0] * scale_factor],
        y=[0, loadings[i, 1] * scale_factor],
        z=[0, loadings[i, 2] * scale_factor],
        mode='lines',
        line=dict(color='black', width=4),
        showlegend=False
    ))
    # Word label
    fig.add_trace(go.Scatter3d(
        x=[loadings[i, 0] * scale_factor],
        y=[loadings[i, 1] * scale_factor],
        z=[loadings[i, 2] * scale_factor],
        mode='text',
        text=[f"<b>{word}</b>"],
        textfont=dict(size=12, color='black'),
        showlegend=False
    ))

fig.update_layout(
    height=1000,
    width=1200,
    autosize=False,
    title_font_size=20
)

fig.show()
```

**Procedural Gravity**

* Observation: Arrows representing terms like `plaintiff`, `defendant`, and `defendants` are clustered in the center, and most of the gray dots are concentrated around them.

* Interpretation: This means that the "standard legal terminology" commonly used in most court rulings forms the center of the data. In other words, ordinary cases fall within the gravitational pull of legal "idioms."

**Semantic Extremes**

* Observation: Arrows for terms like `parody`, `photograph`, and `song` extend towards the bottom and right of the image, and red dots are located on the periphery in those directions.

* Interpretation: These red dots indicate that the documents are not simply long, but rather excessively use words focused on very specific factual patterns, such as "parody" or "photograph copyright." These are cases with strong specificities in particular fields, rather than legal universality.

**Complexity Outliers**

* Observation: Some red dots are seen far away, not aligning with any of the arrows.

* Interpretation: These are truly "mutant" case precedents that use even common legal terms (factor, use) in a non-standard way, or deal with entirely different topics.

```python
explained_variance_3d = pca_3d.explained_variance_ratio_
cumulative_variance_3d = np.cumsum(explained_variance_3d)

print(f"PC1 explains: {explained_variance_3d[0]:.2%} of the variance")
print(f"PC2 explains: {explained_variance_3d[1]:.2%} of the variance")
print(f"PC3 explains: {explained_variance_3d[2]:.2%} of the variance")
print(f"Total variance explained (3D): {cumulative_variance_3d[-1]:.2%}")

plt.figure(figsize=(6, 4))
plt.bar(['PC1', 'PC2', 'PC3'], explained_variance_3d, color='black', alpha=0.7)
plt.title('Variance Explained by 3D Components')
plt.ylabel('Variance Ratio')
plt.show()
```

From 2D to 3D PCA The transition from a 2D to a 3D semantic landscape resulted in a measurable increase in analytical resolution. While the initial 2D projection captured only 4.88% of the variance, the 3D model improved the total explained variance to 6.65% (PC1: 2.65%, PC2: 2.23%, PC3: 1.77%).

Although a total variance of 6.65% remains statistically low, this gain provided the critical "depth" needed to separate Red-flag cases from the dense procedural core. This subtle yet significant increase confirms that legal outliers in the Fair Use dataset are defined by high-dimensional linguistic nuances that cannot be fully captured in a flat 2D plane.
